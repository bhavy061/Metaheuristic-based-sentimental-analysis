```python
from keras.models import Sequential
from keras.layers import Embedding, Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

max_len = len(x_train[0])
vocab_size = 54954
embedding_dim = 100

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len),
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dense(3, activation='softmax')
])

# Compile the Model
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the Model
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# Evaluate the Model
y_pred = model.predict(x_test)
y_pred_classes = y_pred.argmax(axis=1)

# Print Evaluation Metrics
print("Classification Report:\n", classification_report(y_test, y_pred_classes))
print("Accuracy:", accuracy_score(y_test, y_pred_classes))

# Save the Model
model.save("sentiment_cnn_model.h5")
```

```python
import numpy as np
from keras.models import load_model
from pyswarm import pso
from sklearn.metrics import accuracy_score, classification_report

# Load the pre-trained model
model = load_model("sentiment_cnn_model.h5")

# Function to evaluate the model with modified hyperparameters
def evaluate_model(params):
    learning_rate, dropout_rate = params
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Update dropout layer dynamically
    model.layers[4].rate = dropout_rate  # Adjusting the dropout layer rate

    # Make predictions
    y_pred = model.predict(x_test)
    y_pred_classes = y_pred.argmax(axis=1)
    acc = accuracy_score(y_test, y_pred_classes)
    return -acc  # PSO minimizes, so use negative accuracy for maximization

# Defining bounds for hyperparameters
lb = [0.0001, 0.1]  # Lower bounds for learning rate and dropout
ub = [0.01, 0.5]    # Upper bounds for learning rate and dropout

# Performing PSO optimization
best_params, best_score = pso(evaluate_model, lb, ub, swarmsize=10, maxiter=5)

print("Best Hyperparameters (Learning Rate, Dropout Rate):", best_params)
print("Best Accuracy:", -best_score)

# Save the optimized model with the best hyperparameters
model.compile(optimizer=Adam(learning_rate=best_params[0]),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.save("optimized_sentiment_cnn_model.h5")
```

```python
import numpy as np
from keras.models import load_model
from sklearn.metrics import accuracy_score
from skopt import Optimizer
from skopt.space import Real

# Load the pre-trained model
model = load_model("sentiment_cnn_model.h5")

# Define the objective function for ACO
def evaluate_model(params):
    learning_rate, dropout_rate = params
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Adjust the dropout rate
    model.layers[4].rate = dropout_rate

    # Make predictions and calculate accuracy
    y_pred = model.predict(x_test)
    y_pred_classes = y_pred.argmax(axis=1)
    acc = accuracy_score(y_test, y_pred_classes)
    return -acc  # Minimize negative accuracy for maximization

# Set search space for ACO
space = [Real(0.0001, 0.01, name='learning_rate'),
         Real(0.1, 0.5, name='dropout_rate')]

# Initialize ACO Optimizer
optimizer = Optimizer(dimensions=space, random_state=42)

# Run ACO for 10 iterations
for _ in range(10):
    next_point = optimizer.ask()
    val = evaluate_model(next_point)
    optimizer.tell(next_point, val)

# Get the best hyperparameters
best_params = optimizer.Xi[np.argmin(optimizer.yi)]
best_score = -np.min(optimizer.yi)

print("Best ACO Hyperparameters:", best_params)
print("Best Accuracy:", best_score)

# Apply best hyperparameters to the model
model.compile(optimizer=Adam(learning_rate=best_params[0]),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.save("optimized_sentiment_cnn_aco.h5")
```

```python
import numpy as np
from keras.models import load_model
from sklearn.metrics import accuracy_score
from keras.optimizers import Adam

# Define custom GWO optimization class (basic version)
class GWO:
    def __init__(self, fitness_function, lb, ub, pop_size=10, max_iter=10):
        self.fitness_function = fitness_function
        self.lb = np.array(lb)
        self.ub = np.array(ub)
        self.pop_size = pop_size
        self.max_iter = max_iter
        self.alpha = None
        self.beta = None
        self.delta = None
        self.position = None

    def initialize_population(self):
        return np.random.rand(self.pop_size, len(self.lb)) * (self.ub - self.lb) + self.lb

    def update_position(self, positions, a):
        # Update position based on Grey Wolf Algorithm
        r1, r2 = np.random.rand(2)
        A = 2 * a * r1 - a
        C = 2 * r2
        D = np.abs(C * self.alpha - positions)
        new_positions = self.alpha - A * D
        return new_positions

    def solve(self):
        positions = self.initialize_population()
        fitness = np.array([self.fitness_function(p) for p in positions])

        for iteration in range(self.max_iter):
            # Update alpha, beta, delta wolves
            sorted_idx = fitness.argsort()
            self.alpha = positions[sorted_idx[0]]
            self.beta = positions[sorted_idx[1]]
            self.delta = positions[sorted_idx[2]]

            # Update position of all wolves
            a = 2 - iteration * (2 / self.max_iter)
            positions = self.update_position(positions, a)

            # Evaluate fitness
            fitness = np.array([self.fitness_function(p) for p in positions])
            print(f"Iteration {iteration + 1}: Best Fitness = {-min(fitness)}")

        return self.alpha, -min(fitness)  # Return the best solution

# Load the pre-trained model
model = load_model("sentiment_cnn_model.h5")

# Define the fitness function for GWO
def fitness_function(solution):
    learning_rate, dropout_rate = solution
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Adjust dropout dynamically
    model.layers[4].rate = dropout_rate

    # Evaluate model performance
    y_pred = model.predict(x_test)
    y_pred_classes = y_pred.argmax(axis=1)
    acc = accuracy_score(y_test, y_pred_classes)
    return -acc  # Minimize negative accuracy

# Set hyperparameter bounds for GWO
lb = [0.0001, 0.1]  # Lower bounds for learning rate and dropout rate
ub = [0.01, 0.5]  # Upper bounds for learning rate and dropout rate

# Initialize GWO with the custom class
gwo = GWO(fitness_function, lb, ub, pop_size=10, max_iter=10)
best_solution, best_fitness = gwo.solve()

# Display the results
print("Best GWO Hyperparameters:", best_solution)
print("Best Accuracy:", -best_fitness)

# Apply best hyperparameters to the model
model.compile(optimizer=Adam(learning_rate=best_solution[0]),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.save("optimized_sentiment_cnn_gwo.h5")
```

```python
import numpy as np
from keras.models import load_model
from sklearn.metrics import accuracy_score
from keras.optimizers import Adam

# Define custom ACO optimization class
class ACO:
    def __init__(self, fitness_function, lb, ub, pop_size=10, max_iter=10, alpha=1, beta=2, rho=0.5, q0=0.9):
        self.fitness_function = fitness_function
        self.lb = np.array(lb)
        self.ub = np.array(ub)
        self.pop_size = pop_size
        self.max_iter = max_iter
        self.alpha = alpha  # Influence of pheromone
        self.beta = beta    # Influence of heuristic
        self.rho = rho      # Pheromone evaporation rate
        self.q0 = q0        # Exploration vs exploitation balance
        self.pheromone = np.ones(len(self.lb))  # Initial pheromone values

    def initialize_population(self):
        # Initialize the ant population with random solutions
        return np.random.rand(self.pop_size, len(self.lb)) * (self.ub - self.lb) + self.lb

    def choose_move(self, ant_position):
        # Choose the next move based on pheromone and heuristic
        probabilities = np.zeros(len(self.lb))
        for i in range(len(self.lb)):
            probabilities[i] = (self.pheromone[i] ** self.alpha) * ((1 / (self.ub[i] - self.lb[i])) ** self.beta)
        probabilities /= sum(probabilities)
        return np.random.choice(len(self.lb), p=probabilities)

    def update_pheromone(self, ant_positions, fitness_values):
        # Update pheromone values based on the fitness of ants
        best_ant = np.argmin(fitness_values)
        best_position = ant_positions[best_ant]
        best_fitness = fitness_values[best_ant]

        self.pheromone = self.pheromone * (1 - self.rho)  # Pheromone evaporation
        self.pheromone += self.rho * (1 / (1 + best_fitness))  # Add new pheromone

    def solve(self):
        for iteration in range(self.max_iter):
            # Generate ant population
            ant_positions = self.initialize_population()
            fitness_values = np.array([self.fitness_function(p) for p in ant_positions])

            # Update pheromone based on fitness
            self.update_pheromone(ant_positions, fitness_values)

            # Exploration vs exploitation (depending on q0)
            if np.random.rand() < self.q0:
                # Exploitation: focus on best solution
                best_ant = np.argmin(fitness_values)
                best_position = ant_positions[best_ant]
                best_fitness = fitness_values[best_ant]
            else:
                # Exploration: choose a random ant's position
                best_ant = np.argmin(fitness_values)
                best_position = ant_positions[best_ant]
                best_fitness = fitness_values[best_ant]

            print(f"Iteration {iteration + 1}: Best Fitness = {-best_fitness}")

        return best_position, -best_fitness  # Return the best solution

# Load the pre-trained model
model = load_model("sentiment_cnn_model.h5")

# Define the fitness function for ACO
def fitness_function(solution):
    learning_rate, dropout_rate = solution
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Adjust dropout dynamically
    model.layers[4].rate = dropout_rate

    # Evaluate model performance
    y_pred = model.predict(x_test)
    y_pred_classes = y_pred.argmax(axis=1)
    acc = accuracy_score(y_test, y_pred_classes)
    return -acc  # Minimize negative accuracy

# Set hyperparameter bounds for ACO
lb = [0.0001, 0.1]  # Lower bounds for learning rate and dropout rate
ub = [0.01, 0.5]  # Upper bounds for learning rate and dropout rate

# Initialize ACO with the custom class
aco = ACO(fitness_function, lb, ub, pop_size=10, max_iter=10, alpha=1, beta=2, rho=0.5, q0=0.9)
best_solution, best_fitness = aco.solve()

# Display the results
print("Best ACO Hyperparameters:", best_solution)
print("Best Accuracy:", -best_fitness)

# Apply best hyperparameters to the model
model.compile(optimizer=Adam(learning_rate=best_solution[0]),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.save("optimized_sentiment_cnn_aco.h5")
```
